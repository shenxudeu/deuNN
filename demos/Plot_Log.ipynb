```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
import pdb
fn = '/home/xu/Tmp/log.log'
f = open(fn,'r')
cc = 0
train_acc = []
test_acc = []
for line in f.readlines():
    str_list = line.split(',')
    for field in str_list:
        if field[1:10] == 'train_acc':
            tmp = field.split('=')
            train_acc.append(float(tmp[-1]))
        if field[1:8] == 'val_acc':
            tmp = field.split('=')
            test_acc.append(float(tmp[-1]))
```

```python
plt.plot(train_acc)
plt.plot(test_acc)
```
